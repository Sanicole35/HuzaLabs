```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting
plt.plot(x, y, marker='o', linestyle='-', color='y', label='Prime Numbers')

# Customization
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Prime Numbers')
plt.legend()

# Display
plt.show()
```
